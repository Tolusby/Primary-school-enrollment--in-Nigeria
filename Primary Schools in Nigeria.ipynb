# Primary School Enrollment in Nigeria

A brief anaylysis of primary school enrollment gross % in Nigeria with the use of pandas, matplotlib e.t.c. from a xlsx file import.

Data set from OpenAfrica [link](https://africaopendata.org/dataset/school-enrolment/resource/9c600c8a-64b8-4538-a195-4826f3d98208?view_id=c8aac2de-907f-460a-a866-e231d814dfff) by WorldBank

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
headers = ['Year', 'Male', 'Female']
df = pd.read_excel('nigeria-primary-school-enrolment-gross.xlsx', names=headers,
parse_dates=True)
df
```

```python
df.describe()
```

```python
plt.rcParams["figure.figsize"] = [7.50, 5.50]
plt.rcParams["figure.autolayout"] = True


df.set_index('Year').plot()
plt.grid()
plt.show()
```
